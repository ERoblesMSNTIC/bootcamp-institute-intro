```python
# Función para generar datos simulados

def simular_nulos_col(serie, porcentaje=0.1, semilla=None):

    serie_mod = serie.copy()

    if semilla is not None:

        np.random.seed(semilla)

    cantidad = int(len(serie) * porcentaje)

    indices = np.random.choice(serie.index, size=cantidad, replace=False)

    serie_mod.loc[indices] = np.nan

    return serie_mod
```

```python
simular_nulos_col(df.petroleo_bbl)
```

```python
# Describe sin datos nulos
df.petroleo_bbl.describe()
```

```python
# Describe con datos nulos simulados
simular_nulos_col(df.petroleo_bbl).describe()
```

```python
# Describe con datos nulos simulados al 50% de los datos
simular_nulos_col(df.petroleo_bbl, porcentaje=0.5).describe()
```

#### 2. Tratamiento de valores faltantes

```python
# Simular con nuestro df como si tuvieramos valores nulos en algunas columnas
df_missing = df.copy()
df_missing['petroleo_bbl'] = simular_nulos_col(df['petroleo_bbl'],porcentaje=0.2,semilla=1)
df_missing['agua_bbl'] = simular_nulos_col(df['agua_bbl'], porcentaje=0.05,semilla=1)
```

```python
# Método para reasignar valores nulos por medio de interpolación
df_missing.agua_bbl.interpolate(method='linear').describe()
```

#### 3. Detección y manejo de Outliers

- Identificar outliers en presion_cabezal_psi usando el rango intercuartil (IQR)

```python
# Histograma básico
# df['presion_cabezal_psi'].hist()

# Pasar el argumento como una tabla de una sola columna para graficar cajas y bigotes
df['presion_cabezal_psi'].to_frame().boxplot()
```

```python
df['presion_cabezal_psi'].describe()
```

```python
# Paso 1 - Calcular los Cuartiles 1 (25%) y Cuartiles 3 (75%)

Q1 = df['presion_cabezal_psi'].quantile(0.25)
Q3 = df['presion_cabezal_psi'].quantile(0.75)
```

```python
# Paso 2 - Calcular el Rango Intercuartil IQR
IQR = Q3 - Q1
IQR
```

```python
# Paso 3 - Definir los límites del IQR

limite_superior = Q3 + 1.5 * IQR
limite_inferior = Q1 - 1.5 * IQR

(limite_inferior, limite_superior)
```

```python
# Paso 4 - Aplicar función lógica para eliminar outliers

df_presion_limpio = df[ ~(df['presion_cabezal_psi'] < limite_inferior) | (df['presion_cabezal_psi'] > limite_superior)]
```

```python
df_presion_limpio.presion_cabezal_psi.to_frame().boxplot()
```

#### 5. Feature Engineering para datos petroleros

Transformar datos técncos en variables que mejor representen el rendimiento, eficiencia y condiciones operativa de pozos petroleros, facilitando así análisis descriptivos, predictivos y de diagnóstico

Por qué es importante?

- Ayuda a reducir ruido en modelos de predicción (como forecasting de producción)

- Puede resaltar anomalías o problemas operativos (baja eficiencia, exceso de agua)

- Mejora la explicabilidad de los modelos (por qué una predicción fue hecha)

- Facilita la toma de decisiones operativas y de mantenimiento


| **Variable Derivada**     | **Fórmula o Idea Clave**                      | **Interpretación**                                   |
|---------------------------|------------------------------------------------|------------------------------------------------------|
| `eficiencia_extraccion` | petroleo_bbl / horas_operativas             | Producción de petróleo por hora operativa            |
| `gas_agua_ratio`        | gas_mscf / agua_bbl                          | Cantidad de gas producida por unidad de agua         |
| `prod_por_choke`       | petroleo_bbl / choke_percent                | Eficiencia relativa a la apertura del choke          |
| `indice_productividad`  | petroleo_bbl / presion_cabezal_psi          | Producción por unidad de presión                     |
| `eficiencia_termica`    | produccion_equivalente / temperatura_c      | Rendimiento respecto a temperatura                   |
| `indice_trabajo`        | horas_operativas / liquido_total            | Esfuerzo operativo en relación a producción          |
| `ratio_gas_petroleo`    | gas_mscf / petroleo_bbl                      | Proporción gas-petróleo (ya existe en tus datos)     |

```python
# Tasa gas-agua - proporción de gas producido por barril de agua

df['gas_agua_ratio'] = df['gas_mscf'] / df['agua_bbl']
```

```python
# Indice de productividad = Producción por unidad de presión del cabezal

df['indice_productividad'] = df['petroleo_bbl'] / df['presion_cabezal_psi']
```

```python
# Eficiencia térmica (proxy) = Producción por grados Celsius

df['eficiencia_termica'] = df['produccion_equivalente'] / df['temperatura_c']  

# Producción por apertura del choke = rendimiento por porcentaje abierto del choke

df['prod_por_choke'] = df['petroleo_bbl'] / df['choke_percent']
```
